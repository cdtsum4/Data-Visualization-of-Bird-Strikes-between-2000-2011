## Importing Libraries and Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Reading Data

```python
df= pd.read_excel("Bird Strikes data.xlsx", keep_default_na = False)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.columns
```

```python
for column in df.columns:
    df.columns = df.columns.str.replace(r'[^\w\s]','',regex = True)
df.columns
```

#### Renaming Columns and Turning to Lower Case

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace(r'_$', '', regex=True)
df.columns
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

### Data Cleaning 

#### Finding Unique

```python
df['remarks'].unique()
```

```python
# Dropping Table

df.drop('remarks', axis = 1, inplace = True)
```

```python
df['origin_state'].unique()
```

```python
df['aircraft_type'].unique()
```

```python
df['when_phase_of_flight'].unique()
```

```python
df['conditions_precipitation'].unique()
```

```python
df['effect_impact_to_flight'].unique()
```

```python
df.head()
```

```python
# replacing blank values

df = df.replace('', np.nan)
df.isnull().sum()
```

```python
# changing "None" to string

df['conditions_precipitation'] = df['conditions_precipitation'].replace("None", "No Precipitation")
df['effect_impact_to_flight'] = df['effect_impact_to_flight'].replace("None", "No Damage")
df[['conditions_precipitation','effect_impact_to_flight']].head()
```

```python
df.isnull().sum()
```

```python
# makinf copy of data as df1

df1 = df.copy()
df1.head()
```

```python
df1.isnull().any()
```

```python
df1.columns[df1.isnull().any()]
```

```python
df1.isnull().sum()
```

```python
# Dropping  NA values

df1.dropna(subset=['aircraft_type', 'airport_name', 'altitude_bin',
       'wildlife_number_struck', 'effect_impact_to_flight', 'flightdate',
       'aircraft_number_of_engines', 'aircraft_airlineoperator',
       'origin_state', 'when_phase_of_flight',
        'wildlife_size', 'pilot_warned_of_birds_or_wildlife',
       'feet_above_ground', 'is_aircraft_large'], inplace = True)
```
